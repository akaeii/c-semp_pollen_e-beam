```python
import os
import csv
import glob
```

```python
#function for converting dpt files to csv files 

def dpt2csv(dpt_path, main_name, output_name): #dpt file path, searched folder path, folder name for outputs
    csv_name = dpt_path.replace(main_name, output_name).replace('.dpt', '.csv') #name/path of output csv (replaced input path to output path)
    os.makedirs(os.path.dirname(csv_name), exist_ok=True) #creates subdirectories patterned over input subdirectories

    #reformatting (/t to ,) and placing at proper subdirectory in output directory
    with open(dpt_path, 'r', newline='') as dpt_file: 
        read_dpt = csv.reader(dpt_file)
        with open(csv_name, 'w', newline='') as csv_file:
            csv_writer = csv.writer(csv_file)           
            for row in read_dpt:
                for element in row:
                    cleaned = [x for x in element.split('\t')] #splits each element into 2 to as delimited by \t
                    csv_writer.writerow(cleaned)
                    
    print(f'{csv_name} created')
```

```python
#find all csv files in a given directory including subdirectories

main_name = input('• type the main directory to search (must be in the same directory as this notebook): ')
output_name = input('• type the file name of your intended output folder: ')

main_path = os.path.join(str(os.getcwd()), main_name) #appending current working directory with input path
dpt_paths = glob.glob(f'{main_path}/*/*.dpt') #searches subdirectories (1 layer) in input path for dpt files

print(f'{len(dpt_paths)} found')

for dpt_path in dpt_paths:
    dpt2csv(dpt_path, main_name, output_name)
```

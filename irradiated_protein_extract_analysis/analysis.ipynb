```python
import matplotlib.pyplot as plt
import pandas as pd
import glob
%matplotlib widget
```

```python
dose_dirs = glob.glob(f"protein_extract_spectra/*.csv")
dose_names = [0,5,10,20,50]
colors = ["black", "green", "blue", "red", "purple"]

pe0 = pd.read_csv("protein_extract_scan48.csv")
pe10, pe20, pe50, pe5 = [pd.read_csv(dose_dir) for dose_dir in dose_dirs]

wavenumber = pd.read_csv("wave_no.csv")

pe_df_pooled = [pe0, pe5, pe10, pe20, pe50]
for df in pe_df_pooled:
    df['average'] = df.mean(axis=1)
```

```python
pe_ave_fig, axis = plt.subplots(1,1, figsize=(10,5))

for dose_df, dose_name, color in zip(pe_df_pooled, dose_names, colors):
    axis.plot(wavenumber, dose_df["average"],color=color, label=f"{dose_name} kGy")
    axis.legend()
```

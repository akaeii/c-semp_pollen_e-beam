```python
import pandas as pd
import glob
import os
```

```python
def scanning_number_concatenator(search_path_index):

    #directory: main_folder/scanning no. 1/trial 1... trial n, ... scanning no n/

    readings_path = 'spectra_csv' #main folder containing folders of spectra divided into scanning no.
    search_paths = [os.path.join(readings_path, f'{i} Scans') for i in [16, 24, 32, 48, 64, 72]] #list of scanning no. folder names

    scanning_number_csv_paths = glob.glob(f'{search_paths[search_path_index]}/*.csv') #searches for csv files in the selected scanning number folder
 
    print(f'now doing {search_paths[search_path_index]}') #simple tracker for current scanning number being done

          
    sn_trials = [pd.read_csv(path, names=['wave_no', 'abs']) for path in scanning_number_csv_paths] #creates list of dataframes for each trial
    wave_number_standard = pd.DataFrame(sn_trials[0]['wave_no']) #grabs the wave no column of the first trial's dataframe
    
    sn_abs = pd.concat([df['abs'] for df in sn_trials], axis=1, ignore_index=True) #dataframe of concatenated abs data of all trials
    sn_merged = pd.concat([wave_number_standard, sn_abs], axis=1, ignore_index=True) #concatenates wave no column to dataframe of concatenated abs data
    sn_merged.columns = ['wave_no'] + [f'trial_{i}' for i in range(1,len(sn_trials)+1,1)] #renames columns | 'wave no', trial 1, ... trial 15
    
    return sn_merged
```

```python
scanning_number_concatenator(4)
```

```python
scanning_number_concatenator(4).to_csv('64_scans.csv', index=False)
```

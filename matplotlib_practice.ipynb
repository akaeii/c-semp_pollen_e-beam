```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
#super simplified data of atr-ftir analytical range
wavenum_x = [i for i in range(4000, 1000, -500)]
trans_y0 = [0.7, 0.7, 0.9, 1, 1, 0.99]
trans_y1 = [1, 0.7, 0.8, 0.8, 0.9, 1]
```

```python
#styling
plt.style.use('dark_background')

#plotting
plt.plot(wavenum_x, trans_y0,
        label='Sample 0',
        color='#ff3333') #hot red
plt.plot(wavenum_x, trans_y1,
        label='Sample 1',
        color='#52ff33') #bright green

#labelling
plt.title('Sample ATR-FTIR Analytical Range Data')
plt.xlabel('Wave number cm⁻¹')
plt.ylabel('Transmission')

plt.legend()
plt.tight_layout() #fixes padding issues
plt.grid(True)
plt.show() 
```
